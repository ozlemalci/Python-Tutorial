```
s="01234567890123456789"

print("s=",s)
print("s stringinin uzunluğu (len(s))=", len(s))
print("Baştan ilk 4 karaketeri al s[0:4]=",s[0:4])
print("2. karakterden başla 9. sıraya kadar al s[2:9]=",s[2:9])
print("3. karakterden başla sonuna kadar al s[3:]=",s[3:])
print("Son 2 karakter hariç hepsini al s[:-2]=", s[:-2])
print("Son 10 karakterin ilk sekizini al s[-10:-2]=", s[-10:-2])
print("İlk 3 karakterden sonra araya \"abc\" al s[3:]+ \"abc\"+ s[:3] = ",s[:3]+ "abc"+ s[3:])
print("Baştan 2, sondan 4 karakteri al, arayı sil s[0:2]+s[-4:] = ",s[0:2]+s[-4:] )
```
