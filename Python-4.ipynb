```
#lambda fonksiyonu yerel ve global değişkenlere ulaşamayan anonim bir fonksiyondur.

import math
hipotenüs= lambda kenar1,kenar2: math.sqrt(kenar1*kenar1 + kenar2*kenar2)
print(hipotenüs(3,4))
```
